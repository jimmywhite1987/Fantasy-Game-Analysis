```python
final_critic = purchase_data.loc[purchase_data['Item Name'] == 'Final Critic', :]
fc_count = final_critic['Item Name'].count()
fc_price = round(final_critic['Price'].mean(),2)
final_critic2 = final_critic.groupby(['Item Name', 'Item ID']).agg({'Price': sum}).reset_index()
oath = purchase_data.loc[purchase_data['Item Name'] == 'Oathbreaker, Last Hope of the Breaking Storm', :]
oath_count = oath['Item Name'].count()
oath_price = round(oath['Price'].mean(),2)
oath = oath.groupby(['Item Name', 'Item ID']).agg({'Price': sum}).reset_index()
nirvana = purchase_data.loc[purchase_data['Item Name'] == 'Nirvana', :]
nirvana_count = nirvana['Item Name'].count()
nirvana_price = round(nirvana['Price'].mean(),2)
nirvana = nirvana.groupby(['Item Name', 'Item ID']).agg({'Price': sum}).reset_index()
fiery_glass = purchase_data.loc[purchase_data['Item Name'] == 'Fiery Glass Crusader', :]
fiery_count = fiery_glass['Item Name'].count()
fiery_price = round(fiery_glass['Price'].mean(),2)
fiery_glass = fiery_glass.groupby(['Item Name', 'Item ID']).agg({'Price': sum}).reset_index()
# Commenting these two lines out because after reviewing the data, though the purchase count is the same, Extraction has a higher total purchase value!
# persuasion = purchase_data.loc[purchase_data['Item Name'] == 'Persuasion', :]
# persuasion = persuasion.groupby(['Item Name']).agg({'Price': sum}).reset_index()
extraction = purchase_data.loc[purchase_data['Item Name'] == 'Extraction, Quickblade Of Trembling Hands', :]
ex_count = extraction['Item Name'].count()
ex_price = round(extraction['Price'].mean(),2)
extraction = extraction.groupby(['Item Name', 'Item ID']).agg({'Price': sum}).reset_index()
pop_items = pd.concat([final_critic2, oath, nirvana, fiery_glass, extraction], ignore_index=True)
pop_items['Purchase Count'] = [fc_count, oath_count, nirvana_count, fiery_count, ex_count]
pop_items['Item Price'] = [fc_price, oath_price, nirvana_price, fiery_price, ex_price]
pop_items.rename(columns = {'Price':'Total'}, inplace = True)
new_pop = pop_items.set_index('Item Name')
new_pop['Total'] = new_pop['Total'].map("${:.2f}".format)
new_pop['Item Price'] = new_pop['Item Price'].map("${:.2f}".format)
new_pop



# Another Note: I used the mean function for "Final Critic" because it had at least two listed prices and the instructions only ask for one
```

```python
# Since I pulled the item price andpurchase count for 4 of the items, I only have to find the values for the different item, Singed Scalpel
singed = purchase_data.loc[purchase_data['Item Name'] == 'Singed Scalpel', :]
singed_count = singed['Item Name'].count()
singed_price = round(singed['Price'].mean(),2)
```

```python
# These following two lines gave me the 5 most profitable items, so I moved it down and created code above to grab purchase count and item price
profits = purchase_data.groupby(['Item Name', 'Item ID']).agg({'Price': sum}).reset_index()
new_profits = profits.nlargest(5, columns=['Price']).reset_index().head(5)
new_profits['Purchase Count'] = [fc_count, oath_count, nirvana_count, fiery_count, singed_count]
new_profits['Item Price'] = [fc_price, oath_price, nirvana_price, fiery_price, singed_price]
final_profits = new_profits.set_index('Item Name').drop(['index'], axis=1)
final_profits.rename(columns = {'Price':'Total'}, inplace=True)
final_profits['Total'] = final_profits['Total'].map("${:.2f}".format)
final_profits['Item Price'] = final_profits['Item Price'].map("${:.2f}".format)
final_profits
```
